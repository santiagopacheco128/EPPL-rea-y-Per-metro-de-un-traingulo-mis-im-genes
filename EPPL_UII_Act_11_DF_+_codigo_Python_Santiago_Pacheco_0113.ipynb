```python
# Calcular area y perimetro de un triangulo rectangulo
# Variables A, P, b, h, hip
# Formulas A = (b * h) / 2 y P = b + h
# Capturar el valor de h,b
import math
print("Calcular area y perimetro de un triangulo rectangulo")

b = float(input("Introduce el valor de la base (b): "))
h = float(input("Introduce el valor de la altura (h): "))
A = (b * h) / 2

print(f"El area del triangulo es: {A}")
print("Santiago Pacheco NL = 43")
```
